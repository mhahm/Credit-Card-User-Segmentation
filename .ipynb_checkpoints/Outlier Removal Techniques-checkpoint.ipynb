# IQR
One way of detecting and filtering outliers is to user the interquartile range (IQR). The upper limit for our obserations is `Q3 + 1.5 * IQR` and the lower bound is `Q1 - 1.5 * IQR`, where,
* Q1 = 25th percentile 
* Q2 = 75th percentile
* IQR = Q3 - Q1
In other words, the values more extreme than the whiskers of the boxplot

```python
def outliers(df, feature):
    Q3 = df[feature].quantile(0.75)
    Q1 = df[feature].quantile(0.25)
    IQR = Q3 - Q1
    
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR
    
    ls = df.index[(df[feature] < lower_bound) | (df[feature] > upper_bound) ]

    return ls
```

```python
# for loop to run function on multiple features

index_list =[]
for feature in df.columns:
    index_list.extend(outliers(df, feature))
```

# 2nd technique for trimming the outliers:
## capping variables at upper and lower limits
set all the values below the lower bounds equal to the lower i.e. cap the values to the lower bound.
Similarily set all the values above the upper bounds equal to the upper bound

```python
def outliers_2(df, feature):
    Q3 = df[feature].quantile(0.75)
    Q1 = df[feature].quantile(0.25)
    IQR = Q3 - Q1
    
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR
    
    return upper_bound, lower_bound
```

```python
upper_bound, lower_bound = outliers_2(df, feature)

df[df[feature] < lower_bound] = lower_bound
df[df[feature] > upper_bound] = upper_bound
```

# 3rd Technique: Binning aka Discretization
Here we use "equal-width" discretization but you can also use "equal frequency binning"
https://www.geeksforgeeks.org/ml-binning-or-discretization/


```python
maxrange = int(np.ceil(max(df[feature])))
minrange = int(np.floor(min(df[feature])))
featureRange = maxrange - minrange
bins = 10
binwidth = int(np.round(featureRange/bins))
```

```python
intervals = [feature for feature in range(minrange, maxrange+binwidth, binwidth)]
intervals
```

```python
df['feature_bins'] = pd.cut(df[feature], bins=intervals, labels=None)
```

```python
df.groupby('feature_bins')[feature].count().plot.bar()
plt.xticks(rotation=45)
plt.ylabel('observation count')
```
